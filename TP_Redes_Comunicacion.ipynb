# TP - Redes y Comunicación

Crear una red de ocho nodos con relaciones direccionadas donde cada nodo tenga comunicación directa con otros dos nodos.

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
red = nx.DiGraph()

equipos = [
    "Equipo A","Equipo B","Equipo C","Equipo D",
    "Equipo E","Equipo F","Equipo G","Equipo H"
]

for equipo in equipos:
    red.add_node(equipo)
```

```python
enlaces = [
    ("Equipo A", "Equipo D"), ("Equipo A", "Equipo F"),
    ("Equipo B", "Equipo A"), ("Equipo B", "Equipo G"),
    ("Equipo C", "Equipo B"), ("Equipo C", "Equipo H"),
    ("Equipo D", "Equipo C"), ("Equipo D", "Equipo E"),
    ("Equipo E", "Equipo F"), ("Equipo E", "Equipo H"),
    ("Equipo F", "Equipo B"), ("Equipo F", "Equipo G"),
    ("Equipo G", "Equipo D"), ("Equipo G", "Equipo E"),
    ("Equipo H", "Equipo A"), ("Equipo H", "Equipo C")
]

for origen, destino in enlaces:
    red.add_edge(origen, destino)
```

```python
print("Cantidad de nodos:", red.number_of_nodes())
print("Cantidad de conexiones:", red.number_of_edges())
print("Grado de salida de cada nodo:", dict(red.out_degree()))
```

```python
ubicacion = nx.circular_layout(red)

nx.draw(
    red,
    ubicacion,
    with_labels=True,
    node_size=2000,
    font_size=9,
    arrows=True
)

plt.show()
```
